## Loading the data collected

```python
import matplotlib.pyplot as plt
import json
```

```python
f_path = "data_new.json"

# Load the JSON data
with open(f_path) as f:
    data_traffic = json.load(f)

# Take only the 4th onwards
data_traffic = {key:data_traffic[key] for key in list(data_traffic.keys())[4:]}

# Group the data by date
data_traffic_grouped = {}

for date, traffics in data_traffic.items():
    sum = 0
    for traffic in traffics:
        sum += traffic['EstTime']

    data_traffic_grouped[date] = sum

# Print the grouped data
print(data_traffic_grouped)
```

```python
len(data_traffic_grouped)
```

```python
# Plot line graph 
plt.figure(figsize=(20, 10))

# Remove x ticks
plt.xticks([])

# Set x labels
plt.xlabel('DateTime')
plt.ylabel('Total Estimated Time (s)')

# Plot the data
plt.plot(data_traffic_grouped.keys(), data_traffic_grouped.values(), marker='o')
```

```python
# Plot the raw values as a distribution graph based on the occurences
# Get counts
data = list(data_traffic_grouped.values())

import seaborn as sns
sns.displot(data, kde=True, bins=15)
```

```python
# Import tensorflow lstm model
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout
from sklearn.preprocessing import MinMaxScaler
import numpy as np
```

```python
# Convert the data to a numpy array
data_traffic_grouped_np = np.array(list(data_traffic_grouped.values())).reshape(-1, 1)
```

```python
# Normalize the data
scaler = MinMaxScaler()
data_traffic_grouped_np_scaled = scaler.fit_transform(data_traffic_grouped_np)

# Save the scaler
import joblib
joblib.dump(scaler, "scaler.pkl")
```

```python
def create_sequence_no_overlap(data, input_seq_length, output_seq_length):
    X, y = [], []
    # Step size is now 24 (input_seq_length)
    for i in range(0, len(data) - input_seq_length - output_seq_length + 1, input_seq_length):
        X.append(data[i:i+input_seq_length])  # Input sequence (24 hours)
        y.append(data[i+input_seq_length:i+input_seq_length+output_seq_length])  # Output sequence (next 24 hours)
    return np.array(X), np.array(y)

# Use the modified function to create your input-output sequences with no overlap
X, y = create_sequence_no_overlap(data_traffic_grouped_np_scaled, 24, 24)

print("Input shape:", X.shape)  # Should be fewer sequences with no overlap
print("Output shape:", y.shape)
```

```python
# Creating the model using functional
input_x = tf.keras.layers.Input(shape=(24, 1))
x = LSTM(128, activation='relu')(input_x)
x = Dense(24)(x)

model = tf.keras.models.Model(inputs=input_x, outputs=x)

model.compile(optimizer='adam', loss='mse')
```

```python
model.fit(X, y, epochs=30, batch_size=32, validation_split=0.2)
```

```python
# Predict on the training data
train_predictions = model.predict(X)

# Rescale the predictions back to original scale
train_predictions_rescaled = scaler.inverse_transform(train_predictions.reshape(-1, 24))

# Flatten the predictions to get a continuous time series
flattened_predictions = train_predictions_rescaled.flatten()

# Combined the predictions
combined_predictions = np.concatenate((data_traffic_grouped_np[:24].reshape(-1), flattened_predictions), axis=0)

# Convert floats to int
combined_predictions = combined_predictions.astype(int)
```

```python
# Plot the predictions against the actual values
plt.figure(figsize=(20, 10))
plt.plot(range(len(data_traffic_grouped_np)), data_traffic_grouped_np, label='Actual')
plt.plot(range(len(combined_predictions)), combined_predictions,  label='Predicted')
plt.yticks(np.arange(250, 550, 10))
plt.legend()
```
